## No 4

H0: Tidak perbedaan harga rumah yang siginifikan setiap tahunnya <br>
H1: Ada perbedaan harga rumah yang siginifikan setiap tahunnya

```python
import pandas as pd
import numpy as np
import scipy.stats as st

df = pd.read_csv('train.csv')

df['YrSold'].value_counts()
```

```python
tahun1 = df[df['YrSold'] == 2006]
tahun2 = df[df['YrSold'] == 2007]
tahun3 = df[df['YrSold'] == 2008]
tahun4 = df[df['YrSold'] == 2009]
tahun5 = df[df['YrSold'] == 2010]

tahun1['SalePrice'].mean()
tahun2['SalePrice'].mean()
tahun3['SalePrice'].mean()
tahun4['SalePrice'].mean()
tahun5['SalePrice'].mean()

anova_result = st.f_oneway(tahun1['SalePrice'], 
                        tahun2['SalePrice'], 
                        tahun3['SalePrice'], 
                        tahun4['SalePrice'], 
                        tahun5['SalePrice'])

p_value = anova_result.pvalue
print('P-Value:', p_value)
```

```python
if p_value >= 0.05:
    print('terima H0, artinya tidak cukup bukti menyatakan bahwa ada perbedaan harga rumah siginifikan setiap tahunnya')
else:
    print('Terima H1, artinya cukup bukti menyatakan bahwa ada perbedaan harga rumah siginifikan setiap tahunnya')
```

**INTERPRETASI**

Dapat disimpulkan bahwa setiap tahunnya harga rumah itu pasti itu adalah **salah**, karena p-value lebih tinggi dari batas maksimal resiko.

**Pemahaman konsep A/B Testing**
- Apa nama experimentnya? <br>
    A/B Testing pengaruh jumlah campaign terhadap user yang membuka akun deposit 

- Definisikan Hipotesis <br>
    Jumlah campaign berbanding lurus dengan meningkatnya keinginan user untuk membuka akun deposit
    
- Siapa participant-nya? <br>
    Nasabah bank

- Variabel yang akan diuji apa? <br>
    Jumlah campaign

- Metrics apa yang akan digunakan? <br>
    Conversion rate jumlah akun deposit yang terdaftar karena campaign

- Berapa sample size dan durasi experiment?<br>
    Sampel yang dibutuhkan untuk 45211 orang adalah "396"

H0 : Tidak ada perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit dari pengaruh jumlah campaign <br>
H1 : ada perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit dari pengaruh jumlah campaign

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import statsmodels.stats.proportion as sp

df_bank = pd.read_excel('bank-full.xlsx')
df_bank
```

```python
# Mengubah kolom 'y' menjadi nilai biner, 'yes' menjadi 1 dan 'no' menjadi 0
df_bank['y'] = df_bank['y'].map({'yes': 1, 'no': 0})

# Mengelompokkan data berdasarkan 'campaign' dan menghitung rata-rata 'y'
grouped = df_bank.groupby('campaign')['y'].mean()

# Membuat plot
plt.plot(grouped.index, grouped.values)

plt.title('Hubungan antara Jumlah Campaign dan Keinginan User Membuka Akun Deposit')
plt.xlabel('Jumlah Campaign')
plt.ylabel('Rata-Rata Keinginan Membuka Akun Deposit')
plt.show()
```

Berdasarkan histogram diatas, campaign akan dikelompokan menjadi 3 kelompok <= 15, <= 30, dan > 31 

```python
# Membuat dua kelompok berdasarkan 'campaign'
campaign1 = df_bank[df_bank['campaign'] <= 32]  
campaign2 = df_bank[df_bank['campaign'] > 32] 

# Menghitung jumlah sukses untuk setiap kelompok
succes_campaign1 = campaign1['y'].sum()
succes_campaign2 = campaign2['y'].sum()

# Menghitung jumlah observasi untuk setiap kelompok
obs_campaign1 = campaign1.shape[0]
obs_campaign2 = campaign2.shape[0]

# Melakukan uji proporsi z
succes = [succes_campaign1, succes_campaign2]
obs = [obs_campaign1, obs_campaign2]

stats, pvalue = sp.proportions_ztest(succes, obs)

print('P-Value:', pvalue)
```

```python
if pvalue >= 0.05:
    print('Terima H0, artinya tidak ada perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit dari pengaruh jumlah campaign')
else:
    print('Terima H1, artinya ada perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit dari pengaruh jumlah campaign')
```

**Pengujian Hipotesis**

**a. Penentuan Hipotesis:**
   - **Hipotesis Nol (H0)**: Tidak ada perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit antara kelompok yang mengikuti campaign dan kelompok yang tidak mengikut campaign.
   - **Hipotesis Alternatif (H1)**: Terdapat perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit antara kelompok yang mengikuti campaign dan kelompok yang tidak mengikut campaign.

**b. Pengujian yang Digunakan:**
   - Dalam pengujian ini menggunakan uji proporsi z untuk membandingkan dua kelompok berdasarkan jumlah keberhasilan dan jumlah observasi. Uji proporsi z adalah metode statistik yang sesuai untuk menguji apakah ada perbedaan yang signifikan antara dua proporsi (dalam hal ini, proporsi keberhasilan dalam membuka akun deposit).

**c. Deteksi Asumsi beserta Cara Penanganannya:**
   - Dalam uji proporsi z, asumsi utama adalah bahwa data mengikuti distribusi binomial. Untuk data yang cukup besar, seperti yang umumnya terjadi dalam eksperimen A/B Testing, asumsi ini biasanya terpenuhi.
   - Tidak ada asumsi lain yang perlu dipertimbangkan dalam uji proporsi z.

**d. Hasil Pengujian Hipotesis:**
   - Hasil pengujian hipotesis menunjukkan bahwa nilai p-value adalah **0.024**. Dalam kasus Anda, jika Anda mendapatkan nilai p-value yang kurang dari batas maksimal resiko **0.05**, Dengan demikian dapat simpulkan bahwa terdapat perbedaan yang signifikan antara dua kelompok.

**e. Kesimpulan dan Interpretasi:**
   - Ini berarti ada bukti statistik yang cukup untuk mendukung hipotesis alternatif (H1), yang menunjukkan bahwa terdapat perbedaan yang signifikan dalam keinginan user untuk membuka akun deposit antara kelompok yang mengikuti campaign dan kelompok yang tidak mengikut campaign.
